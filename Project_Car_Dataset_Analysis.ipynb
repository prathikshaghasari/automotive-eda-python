# **1. Data Cleaning**

### **1.1 Remove duplictes**

```python
df_car = df_car.drop_duplicates()
```

### **1.2 Handle Missing Data**

Here are the columns in the dataset with missing values:
*   Engine HP:  **69** missing

*   Engine Cylinders:  **30** missing
*   Engine Fuel Type:  **3** missing
*   Number of Doors:  **6** missing
*    Market Category:  **3,742** missing












```python
# Filling numerical columns with mean/median

# Engine HP: Filling with the mean (due to right skewed distribution)
df_car['Engine HP'] = df_car['Engine HP'].fillna(df_car['Engine HP'].mean())

# Engine Cylinders: Filling with median (data has outliers, hence medain is prefered)
df_car['Engine Cylinders'] = df_car['Engine Cylinders'].fillna(df_car['Engine Cylinders'].median())

```

```python
# Filling categorical columns with mode

# Engine fuel type: Filling with mode is quite natural
df_car['Engine Fuel Type'] = df_car['Engine Fuel Type'].fillna(df_car['Engine Fuel Type'].mode()[0])

# Number of Doors: Filling with mode likely represents most common configuration and minimizes distortion
df_car['Number of Doors'] = df_car['Number of Doors'].fillna(df_car['Number of Doors'].mode()[0])
```

```python
# Market Category: Too many missing values (almost half of dataset), so dropping rows would lose a lot of data.
df_car['Market Category'] = df_car['Market Category'].fillna('Unknown')
```

```python
# Confirm no missing values remain
df_car.isnull().sum()
```

## **2. Feature Engineering**

Create new, meaningful features from the cleaned dataset to enhance analysis and model building.

**2.1  Calculate "Total MPG" by averaging the "City MPG" and "Highway MPG" columns.**

```python
# Calculate "Total MPG" as the average of city mpg and highway MPG
df_car_filtered.loc[:, 'Total MPG'] = (df_car_filtered['city mpg'] + df_car_filtered['highway MPG']) / 2
```

```python
# Display the first few rows to confirm the new column
df_car_filtered[['city mpg', 'highway MPG', 'Total MPG']].head()
```

**2.2  Compute "Price per HP" by dividing the "MSRP" (Manufacturer's Suggested Retail Price) by the "Engine HP" (Horsepower) column**



```python
# Compute "Price per HP" by dividing MSRP by Engine HP
df_car_filtered.loc[:, 'Price per HP'] = df_car_filtered['MSRP'] / df_car_filtered['Engine HP']
```

```python
# Display a few rows to confirm the new column
df_car_filtered[['MSRP', 'Engine HP', 'Price per HP']].head()
```

**5. A line plot showing trends in average city mpg and highway MPG for different Transmission Type.**


```python
# Calculate average city mpg and highway MPG for each Transmission Type
avg_mpg_by_trans = df_car_filtered.groupby('Transmission Type')[['city mpg', 'highway MPG']].mean().reset_index()

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(avg_mpg_by_trans['Transmission Type'], avg_mpg_by_trans['city mpg'], marker='o', label='City MPG')
plt.plot(avg_mpg_by_trans['Transmission Type'], avg_mpg_by_trans['highway MPG'], marker='o', label='Highway MPG')
plt.title('Average City MPG and Highway MPG by Transmission Type')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

**Interpretation:**

*   Continuously Variable Transmissions (CVT) tend to offer the best fuel efficiency (highest MPG), both in city and highway driving.
*   Manual transmissions also perform well, especially in city MPG  possibly due to more efficient driving control.
*   Automatic transmissions show slightly lower MPG on average compared to CVT and Manual types.
*   The gap between city and highway MPG is consistent across types, with highway MPG always higher.





## **3.4 Correlation Analysis**


Investigate the correlation between the following variables:

*   Engine HP
*   MSRP
*   Popularity
*   city mpg
*   highway MPG




```python
correlation_data = df_car_filtered[['Engine HP', 'MSRP', 'Popularity', 'city mpg', 'highway MPG']]

# To Compute correlation matrix
correlation_matrix = correlation_data.corr()
print(correlation_matrix)

# To Display the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm_r', linewidths=0.5)
plt.title('Correlation Matrix of Selected Variables')
plt.show()
```

**Interpreation**:


 The correlation analysis among Engine HP, MSRP, Popularity, City MPG, and Highway MPG reveals several meaningful patterns.

*  Engine HP and MSRP exhibit a strong positive correlation (~0.70). This suggests that as a vehicle's horsepower increases, its price tends to rise accordingly. This is expected, as high-performance vehicles with more powerful engines are typically positioned in premium or luxury segments.

*   In contrast, Engine HP shows a strong negative correlation with fuel efficiency, especially with City MPG (~ -0.75) and Highway MPG (~ -0.70). This indicates that more powerful engines generally consume more fuel, resulting in lower mileage. The trade-off between performance and efficiency is a well-known trend in automotive design.
*   City MPG and Highway MPG are very strongly correlated (~0.95). This near-perfect relationship means that vehicles efficient in city driving are also efficient on highways. It reflects consistent engineering practices where overall fuel economy improves uniformly across conditions.


*  MSRP and MPG show a moderate negative correlation (~ -0.45 to 0.55). While more expensive vehicles often have lower mileage due to their power and weight, this trend is less rigid than that between HP and MPG. Some high-priced vehicles, particularly hybrids and electric models, still offer excellent fuel efficiency.

*   Finally, Popularity displays very weak correlations (near 0) with all other variables. This implies that a car's market popularity is largely independent of its power, price, or efficiency. Factors such as brand loyalty, aesthetic appeal, marketing, and perceived reliability may play a much larger role in determining popularity.












# **Conclusion**

**Key Insights from the Automotive Market Analysis**


1.   Vehicle Size and Price: There is a clear positive relationship between vehicle size and price.
Larger vehicles (e.g., midsize and full-size) tend to have higher MSRPs, aligning with their increased performance, capacity, and perceived value.
In contrast, compact cars are the most affordable and dominate in terms of sales volume, making them attractive to price-sensitive consumers.

2.   Engine Horsepower and Price: A strong positive correlation between Engine HP and MSRP indicates that more powerful engines are commonly found in higher-priced vehicles, particularly in luxury or performance segments.



3. Transmission Type and Fuel Efficiency-Transmission type significantly impacts fuel efficiency: Direct drive systems offer superior city MPG,
While manual and automatic transmissions perform better on highways.
This differentiation helps manufacturers align transmission systems with use-case preferences.


4. Driven Wheels and Vehicle Pricing: Vehicles with All-Wheel Drive (AWD) and Rear-Wheel Drive (RWD) configurations tend to have higher average prices, reflecting their enhanced performance, traction, and utility value, especially in performance or off-road vehicles.

5. Market Strategy (Size vs. Volume): As observed in the pricing vs. size graph, compact vehicles offer affordability and dominate in volume, making them ideal for mass-market targeting.
Stakeholders can leverage this insight to align offerings with consumer affordability preferences and prioritize market segments with high sales potential.

6. Popularity and Non-Technical Factors: Popularity does not correlate strongly with technical specifications such as engine power or price.
Instead, it is driven by branding, consumer trust, and perception.
To influence sales, stakeholders should continue investing in marketing, brand equity, and customer engagement strategies, which can significantly shape buying decisions.
